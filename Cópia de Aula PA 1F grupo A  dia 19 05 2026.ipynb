```python
1# Escrever um programa que recebe as notas dos 04 bimestre e imprimi a media com o
# Resultando de aprovado, recuperação ou reprovado.

# Regras
# Utilizar while
# Nota igual ou maior que 7 aprovado
# Nota igual ou maior que 5 recuperação
# Nota igual ou menor que 5 reprovado.

soma_notas = 0
bimestre = 1

while bimestre <= 4:
    nota = float(input(f"Digite a nota do {bimestre}º bimestre: "))
    soma_notas += nota
    bimestre += 1

media = soma_notas / 4
print(f"Média Final: {media}")

if media >= 7:
    print("Resultado: Aprovado")
elif media >= 5:
    print("Resultado: Recuperação")
else:
    print("Resultado: Reprovado")
```

```python
saldo = 1000.0

while True:
    print("\n1. Saldo | 2. Depositar | 3. Sacar | 4. Sair")
    opcao = input("Opção: ")

    if opcao == "1":
        print(f"Saldo: R$ {saldo:.2f}")

    elif opcao == "2":
        valor = float(input("Valor do depósito: "))
        if valor > 0:
            saldo += valor
            print("Depósito realizado.")

    elif opcao == "3":
        valor = float(input("Valor do saque: "))
        if 0 < valor <= saldo:
            saldo -= valor
            print("Saque realizado.")
        else:
            print("Saldo insuficiente ou valor inválido.")

    elif opcao == "4":
        print("Até logo!")
        break

```
