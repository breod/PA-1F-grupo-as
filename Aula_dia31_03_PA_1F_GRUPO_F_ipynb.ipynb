```python
# Em um jog, um personagem possui uma variavel chamada energia. O programa deve:
# Se a energia for maior que 70 - Forte, se a energia estiver entre 30 e 70 - Cansand,
# se energia estiver menor que 30 - Fraco

# Entrada

Energia = int (input("digite a energia do seu personagem"))

#Processamento / Saida

if Energia <= 30:
  print (f" Seu personagem tem energia igual a {Energia}. Ele esta fraco.")

elif Energia <= 70:
  print (f" Seu personagem tem energia igual a {Energia}. Ele esta cansado.")

else:
  print (f" Seu personagem tem energia igual a {Energia}. Ele esta forte.")
```

```python
# Estrutura de decisões

# Escrever um programa que recebe o nome e a idade do usúario e
# verifica se o usúario é criança, adolescente e adulto

# <= 12 anos é criança
# <= 17 anos é adolescente
# Adulto

# Entrada
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: ")) # Convert idade to int

# Processamento

if idade <= 12:
    print(f" {nome}, voce tem {idade} anos de idade, voce é criança. ")
elif idade <= 17:
    print(f" {nome}, voce tem {idade} anos de idade, voce é adolescente. ")
else:
    print(f" {nome}, voce tem {idade} anos de idade, voce é adulto. ")
```

```python
# Escrever um programa que classifica os tipos de triangulo e valida se é um triangulo

# Validar se lados digitados pode forma um triangulo
# Lado1 + Lado2 > Lado3 and Lado1 + Lado3 > Lado2 and Lado2 + Lado3 > Lado1

# Tipos de triangulo
# Equilatero - todos os lados iguais - ==
# Escaleno - Todos os lados diferentes - !=
# Isoceles - dois lados iguais

# Entrada

lado1 = int(input("digite o lado 1 do triangulo: "))
lado2 = int(input("digite o lado 2 do triangulo: "))
lado3 = int(input("digite o lado 3 do triangulo: "))

# Verifica se é um triangulo válido
if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    if lado1 == lado2 == lado3:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo equilátero.")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo isósceles.")
    else:
        print(f"Os lados {lado1}, {lado2} e {lado3} formam um triângulo escaleno.")
else:
    print(f"Os lados {lado1}, {lado2} e {lado3} não formam um triângulo válido.")
```
