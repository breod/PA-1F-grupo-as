```python
from re import MULTILINE
# Operadores aritméticos
# Programa que calcula operações com 02 números

# Entrada
a = int(input("Digite um número: "))
b = int(input("Digite outro número: "))

# Processamento
soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_Int = a // b
resto = a % b
potencia = a ** b
# Saída
print("\n--------------OPERADORES ARITMÉTICOS--------")
print("\n--------------SOMA----------")
print(f"O resultado da SOMA de {a} MAIS {b} é igual a: {soma}")
print("--------------------------------------------")
print("\n--------------SUBTRACAO----------")
print(f"O resultado da SUBTRACAO de {a} MENOS {b} é igual a: {subtracao}")
print("--------------------------------------------")
print("\n--------------MULTIPLICACAO----------")
print(f"O resultado da MULTIPLICACAO de {a} VEZES {b} é igual a: {multiplicacao}")
print("--------------------------------------------")
print("\n--------------DIVISAO---------")
print(f"O resultado da DIVISAO de {a} DIVIDIDO {b} é igual a: {divisao}")
print("--------------------------------------------")
print("\n--------------DIVISAO INTEIRA---------")
print(f"O resultado da DIVISAO INTEIRA de {a} DIVIDIDO {b} é igual a: {div_Int}")
print("--------------------------------------------")
print("\n-------------- RESTO DA DIVISAO-----------")
print(f"O resultado da DIVISAO de {a} DIVIDIDO {b} o RESTO da divisao é igual a : {resto}")
print("--------------------------------------------")
print("\n-------------- POTENCIAÇÃO-----------")
print(f"O resultado da POTENCIAÇÃO de {a} ELEVADO {b} é igual a : {potencia}")
print("--------------------------------------------")
```

```python
# Fazer um programa para converter a temperatura
# F= (C x 9/5) = + 32

# Entrada

celsius = float(input("Digite a temperatura em Celsius: "))

# Processamentos

# Comparação

# Aplica a fórmula: F = (C * 9/5) + 32

fahrenheit = (celsius * 9/5) + 32

# Saída

print(f"{celsius}°C equivale a {fahrenheit}°F")
```
